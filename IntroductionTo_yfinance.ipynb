# Stock/Options/Fundamental Data with Ticker object of yfinance

```python
import pandas as pd
import yfinance as yf
```

```python
symbol = 'AAPL'
ticker = yf.Ticker(symbol)
```

```python
df = ticker.history(start='2021-01-01', end='2022-01-01')
df
```

Daily price of stock of Apple

```python
df = ticker.history(start='2021-01-01', end='2022-01-01', actions=False, rounding=True)
df
```

Increase granularity of stock price to 1 minute

```python
df = ticker.history(period='max', actions=False, rounding=True, interval='1m')
df
```

Only 7 days worth of 1m granularity data are allowed to be fetched per request

```python
df = ticker.history(period='5d', actions=False, rounding=True, interval='1m', prepost=False)
df
```

info() to get information about the dataframe returned by yfinance API

```python
df.info()
```

Setting the prepost = True

```python
df = ticker.history(period='5d', actions=False, rounding=True, interval='1m', prepost=True)
df
```

Returns information about stock, pre and post market hours

```python
df.info()
```

## Market data for Multiple stocks

```python
symbols = 'AAPL META GOOG'
ticker = yf.Tickers(symbols)
```

We get a multi-index dataframe

```python
df = ticker.history(interval='1d', start='2023-01-01', actions=False, rounding=True)
df
```

Extract only AAPL and META tickers for Open

```python
idx = pd.IndexSlice
df.loc[:, idx['Close',['AAPL','META']]]
```

Let's say we only want the Closing price as a single index

```python
df['Close']
```

```python
df['Close'][['AAPL','META']]
```

using groupeby parameter

```python
df = ticker.history(interval='1d', start='2023-01-01', actions=False, rounding=True, group_by='tickers')
df
```
